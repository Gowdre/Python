```python
def info(s):
    v='aeiou'
    d={}
    for w in s.split():
        rev=w[::-1]
        vc=sum(1 for ch in w if ch in v)
        even=w[::2]
        d[w]=[rev,vc,even]
    return d
print(info('kabab is love'))
```

```python
def kabab(s):
    out={}
    for i in s.split():
        c=0
        for j in i:
            if j in 'aeiouAEIOU':
                c+=1
        out[i]=[i[::-1],c,i[::2]]
    return out
print(kabab('kabab is love'))
```

```python
# wap factorial using 4th function ##
def fact(n):
    if n==0 :
        return 1
    return n * fact(n-1)
print(fact(5))
```

```python
def fact(n):
    fact=1
    for i in range(1,n+1):
        fact*=i
    return fact
```

```python
def is_strong(num):
    temp=num
    summ=0
    while num!=0:
        d=num%10
        summ+=fact(d)
        num//=10
    if temp==summ:
        return True
    
is_strong(145)
        
```

```python
def arms(n):
    s=0
    temp=n
    c=len(str(n))
    while n>0:
        d=n%10
        s=s+d**c
        n//=10
    if s==temp:
        print('armstrong')
    else:
        print('not armstrong')
arms(2)
```
