```python
l=['hello',745,79.5,63,52552,14.5,'bye']
out =[]
for i in l:
    if type(i)==int:
        rev=0
        while i!=0:
            d=i%10
            rev=(rev*10)+d
            print('rev',rev)
            i=i//10
        out.append(rev)
    else:
        out.append(i)
print(out)
```
